```python

import pandas as pd
a = pd.read_csv("data.csv")
print(a)
head = a.head(5)
print("\n first 5 rows of the dataset is\n", head)
tail =a.tail(5)
print("\n last five rows of the dataset is \n",tail)
describe = a.describe()
print("\n the data description is \n\n",describe)
sum = a.sum()
print("\n the sum of each column is \n",sum)
max = a.max()
print("\n the maximum value of each column is \n",max)
min = a.min()
print("\n the minimum value of each column is \n",min)
data1 = a[["bedrooms", "bathrooms"]]
standard_deviation = data1.std()
print("\n the standard deviation of price and bedrooms is \n",standard_deviation)
correlation = data1.corr()
print("\n the correlation of price and bedrooms is \n",correlation)
count = a.count()   
print("\n the count of each column is \n",count)
skewness = a.select_dtypes(include='number').skew()
print("\n the skewness of each column is \n\n",skewness)
is_null = a.isnull()
print("\n the number of null values in each column is \n\\n",is_null)
fillna = a.fillna(0)
print("\n the dataset after filling the null values with 0 is \n\n",fillna) 
iloc = a.iloc[0:5,0:3]
print("\n the first five rows and first three columns of the dataset is \n\n",iloc)





```
